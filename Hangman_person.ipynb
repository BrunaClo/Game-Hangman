```python
import os 

def welcome_player():
    name=input("Hey, what's your name? ").capitalize()
    print(f"\nHello and Welcome {name}!")
    print(f"\n{'*'*10} Let's Play Hangman {'*'*10}")
    name_user = name
    return name_user

import random
def picking_file():
    """for acces list of word (ALL CAPS):
      http://www.mieliestronk.com/wordlist.html
      
    """ 
    print(f"\nPlease wait : Updating files...")
    word_list = open('word_list.txt', 'r')
    return word_list

def random_valid_word(word_list):
    word_possible = word_list.readlines()
    for i in word_possible:
        if len(i) > 9:
            word_possible.remove(i)
    word_pc=random.choice(word_possible)
    return word_pc


def play_game(word_pc, name_user):
    word_to_guess = "-" *(len(word_pc)-1)
    rounds = 8    
    used_letter = set()
    used_word = set()
    win = False
    print(word_pc)
    print(len(word_pc))
    print(display_hangman(rounds))
    while win == False and rounds > 0:
        choice_user = input("Please try a letter or a word : ").upper()
        if used_letter != set():
            print(f"\tThe letters already guessed are : {used_letter}")
        if used_word != set():
            print(f"\tThe letters already guessed are : {used_word}")
        print(word_pc)
        print(len(word_pc))
        print(type(word_pc))
        print(len(choice_user))
        print(type(choice_user))
        if len(choice_user) == 1 and choice_user.isalpha():
            if choice_user in used_letter:
                print(f"You already try this letter : {choice_user}")
                rounds += 0
            elif choice_user not in word_pc:
                print(f"{choice_user} has no place in this word")
                used_letter.add(choice_user)
                rounds -= 1
            else:
                print(f"\nCool, {choice_user} is in the word")
                used_letter.add(choice_user)

                word_to_complete_list = list(word_to_guess)
                indices = [n for n, l in enumerate(word_pc) if l == choice_user]
                for index in indices:
                    word_to_complete_list[index] = choice_user
                word_to_guess = "".join(word_to_complete_list)
                if word_to_guess in word_pc:
                    win = True             
            
        elif len(choice_user) == len(word_pc)-1 and choice_user.isalpha():
                if '-' not in word_to_guess:
                    print(f"\nUhuul! {choice_user} is the correct word!")
                    word_to_guess = choice_user  
                    win = True                    
                elif choice_user in used_word:
                    print(f"\nYou already try this word : {choice_user}")
                    print(f"""\nThese are the words already chosen :
                    {used_word}""")
                else:
                    print(f"\n{choice_user} is not the word. Sorry :(")
                    used_word.add(choice_user)
                    rounds -= 1                   
        else:
            print(f"\nWhat are you doing? {choice_user} it's not even a possibility")
        print(display_hangman(rounds))
        print(f"\t\t\t\t{word_to_guess}")
            
    if win == True:
        print(f"{word_to_guess} is the right {word_pc}")
        print(f"\n{name_user} saved the hanging man! {escaped_man} ")
    else:
        print(f"""\n{name_user} just let this man die!
        The word to save him was {word_pc}""")   

escaped_man= """
                               O_/ < FREEDOOMMMM
                             _/|
                             __)\\
                             `  /_
                        
            """ 
                       

def display_hangman(rounds):
    stages = [  """
    
                   -------+
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |    ./ \\.
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     /
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |      |
                   |      |
                   |
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |      |
                   |      
                   |
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |
                   |
                   |
                  ===
                   """,
                   """
                   
                   -------+
                   |      |
                   |      
                   |
                   |
                   |
                  ===
                   """
    ]


    return stages[rounds]

def re_play(hangman):
    try_again = False
    while not try_again:
            re_play_it = (input(f"\nWould you like to re-play it?").upper())
            if re_play_it.isalpha():
                if re_play_it == 'YES' or re_play_it == 'Y':
                    try_again = True
                    print(hangman())
                elif re_play_it == 'NO' or re_play_it == 'N':
                    print(f"\nWell, in France do we prefer to cut the heads off")
                    break
            else:
                    print(f"{re_play_it} is not a valid option. Please choose Yes or No")
                    try_again = False     
            
def hangman(): 
        name_user = welcome_player()
        word_list = picking_file()
        word_pc = random_valid_word(word_list)
        play_game(word_pc, name_user)
        re_play(hangman)
 
  
```
