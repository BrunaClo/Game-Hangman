# HANGMAN

```python
def welcome_player():
    name=input("Hey, who will be playing? ").capitalize()
    print(f"\nHello and Welcome {name}!")
    print(f"\n{'*'*10} Let's Play Hangman {'*'*10}")
    name_user = name
    return name_user
```

```python
def picking_file():
        """ Define a function to access the text file downloaded (word_list.txt)
        with (open 'r') and return the word_list to randomize it
        Depending on the size of the file, the function (jupyter) may take a,
        while to charge
      
        for acces the list of word (ALL CAPS) used here:
          http://www.mieliestronk.com/wordlist.html 
        """
        print(f"""\nUpdating files...\n
        Please wait while we choose a word :)""")
        word_list = open('word_list.txt', 'r')
        return word_list
```

```python
import random

def random_valid_word(word_list):
    """Define a function to make each word from the word_list
    an object to a list with (readlines()),
    then randomize the words that have less than 9 differrents
    letter and return the word chosen by the computer
    """
    word_possible = word_list.readlines()
    for i in word_possible:
        if len(i) > 9:
            word_possible.remove(i)
    word_pc=random.choice(word_possible)
    return word_pc.strip("\n")
```

```python
escaped_man= """
                               O_/ < FREEDOOMMMM
                             _/|
                             __)\\
                             `  /_
                        
            """ 
                       

def display_hangman(rounds):
    stages = [  """
    
                   -------+
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |    ./ \\.
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     /
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |      |
                   |      |
                   |
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |      |
                   |      
                   |
                  ===
                   """,
                   """
                   -------+
                   |      |
                   |      O
                   |
                   |
                   |
                  ===
                   """,
                   """
                   
                   -------+
                   |      |
                   |      
                   |
                   |
                   |
                  ===
                   """
    ]


    return stages[rounds]
```

```python
def re_play(hangman):
    try_again = False
    while not try_again:
        re_play_it = (input(f"\nWould you like to re-play it? ").upper())
        if re_play_it.isalpha() and re_play_it == 'YES' or re_play_it == 'Y':
            try_again = True
            print(hangman())
        elif re_play_it.isalpha() and re_play_it == 'NO' or re_play_it == 'N':
            print(f"\nWell, in France we do prefer to cut the heads off...")
            break
        else:
            if " " in re_play_it:
                print(f"\nI can't see your answer")
            else:
                print(f"{re_play_it} is not a valid option. Please choose Yes or No.")
            try_again = False     
```

```python
def lives(rounds):
    if rounds == 8:
        live = print(f"And you have {rounds} tries")
    elif rounds == 1:
        live = print(f"\nBrace yourself, you only got {rounds} try left")
    else:
        live = print(f"\nYou still have {rounds} tries.")
    return live
```

```python
def play_game(word_pc, name_user):
    word_to_guess = "-"*(len(word_pc))
    rounds = 8  
    used_letter = set()
    used_word = set()
    win = False
    print(f"The word has {len(word_pc)} letters")
    while win == False and rounds > 0:
        lives(rounds)
        print(display_hangman(rounds))
        choice_user = input("Please try a letter or a word : ").upper()
        if used_letter != set():
            print(f"\tThe letters already guessed are : {used_letter}")
        if used_word != set():
            print(f"\tThe letters already guessed are : {used_word}")
            
        if len(choice_user) == 1 and choice_user.isalpha():
            if choice_user in used_letter:
                print(f"\nYou already try this letter : {choice_user}")
            elif choice_user not in word_pc:
                print(f"\n{choice_user} has no place in this word...")
                used_letter.add(choice_user)
                rounds -= 1
            else:
                print(f"\nCool, {choice_user} is in the word")
                used_letter.add(choice_user)
                word_to_complete_list = list(word_to_guess)
                indices = [n for n, l in enumerate(word_pc) if l == choice_user]
                for index in indices:
                    word_to_complete_list[index] = choice_user
                word_to_guess = "".join(word_to_complete_list)
                if word_to_guess in word_pc:
                    win = True         
            
        elif len(choice_user) == len(word_pc) and choice_user.isalpha():
            if choice_user == word_pc:
                print(f"\nUhuul! Good guess!")
                word_to_guess = choice_user  
                win = True                   
            elif choice_user in used_word:
                print(f"\nYou already try this word : {choice_user}")
            else:
                print(f"\n{choice_user} is not the word. Sorry :(")
                used_word.add(choice_user)
                rounds -= 1               
        else:
            if " " in choice_user:
                print(f"\nWith blank spaces {name_user} won't go far...")
            else:
                print(f"\nWhat are you doing? {choice_user} it's not even a possibility!")
        print(f"\t\t\t\t\t\t{word_to_guess}")
    if win == True:
            print(f"{word_to_guess} is the right {word_pc}!")
            print(f"\n{name_user} saved the hanging man! {escaped_man}")
    else:
            rounds = 0
            print(f"\n{name_user} just let this man die! {display_hangman(rounds)}")
            print(f"\nThe word to save him was {word_pc}!")
```

```python
def hangman(): 
        name_user = welcome_player()
        word_list = picking_file()
        word_pc = random_valid_word(word_list)
        play_game(word_pc, name_user)
        re_play(hangman)
```
